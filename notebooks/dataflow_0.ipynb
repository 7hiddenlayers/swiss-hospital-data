```python
from dataflows import Flow, load, printer

def drop_invalid_rows(rows):
    for row in rows:
        if row['JAHR'] == 2017 and \
           row['Inst'] != 'Ganze Schweiz':
                yield row

Flow(
    load('../src/kzp17_daten.xlsx', sheet='KZ2017_KZP17'),
    drop_invalid_rows,
    printer(num_rows=1, tablefmt='html')
).results()[0]
```
